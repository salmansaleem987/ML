```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[: , 3:].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []                                                           
for i in range(1, 11):
    model=KMeans(n_clusters=i , init='k-means++', random_state=42)  
                                                                    
    model.fit(X)
    wcss.append(model.inertia_)                                     
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=5 , init='k-means++', random_state=42)
y_kmeans= model.fit_predict(X)
```

```python
print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans==0, 0],X[y_kmeans == 0, 1], s=100, c= 'skyblue', label= 'Cluster1')
plt.scatter(X[y_kmeans==1, 0],X[y_kmeans == 1, 1], s=100, c= 'blue', label= 'Cluster2')
plt.scatter(X[y_kmeans==2, 0],X[y_kmeans == 2, 1], s=100, c= 'green', label= 'Cluster3')
plt.scatter(X[y_kmeans==3, 0],X[y_kmeans == 3, 1], s=100, c= 'gray', label= 'Cluster4')
plt.scatter(X[y_kmeans==4, 0],X[y_kmeans == 4, 1], s=100, c= 'lightgreen', label= 'Cluster5')
plt.scatter(model.cluster_centers_[: ,0], model.cluster_centers_[:,1] ,s=200, c='red', label='centroids')
plt.title('Cluster of Customers')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
