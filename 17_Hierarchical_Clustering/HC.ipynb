```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[: , 3:].values
```

```python
X
```

```python
plt.figure(figsize=[15,15])
import scipy.cluster.hierarchy as sch                       
dendrogram = sch.dendrogram(sch.linkage(X ,method='ward'))   
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=5 ,affinity='euclidean', linkage='ward')
y_hc = model.fit_predict(X)
print(y_hc)
```

```python
plt.scatter(X[y_hc==0, 0],X[y_hc == 0, 1], s=100, c= 'skyblue', label= 'Cluster1')
plt.scatter(X[y_hc==1, 0],X[y_hc == 1, 1], s=100, c= 'blue', label= 'Cluster2')
plt.scatter(X[y_hc==2, 0],X[y_hc == 2, 1], s=100, c= 'green', label= 'Cluster3')
plt.scatter(X[y_hc==3, 0],X[y_hc == 3, 1], s=100, c= 'gray', label= 'Cluster4')
plt.scatter(X[y_hc==4, 0],X[y_hc == 4, 1], s=100, c= 'lightgreen', label= 'Cluster5')
plt.title('Cluster of Customers')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
